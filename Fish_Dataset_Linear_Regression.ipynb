```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('Fish.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# menampilkan berat ikan 0
df[df.Weight == 0]
```

```python
# menghapus jenis ikan dengan berat 0
df = df.drop(df[df.Weight == 0].index)
```

```python
df.shape
```

```python
# menampilkan jenis species ikan
df.Species.unique()
```

```python
df.Species.value_counts()
```

```python
# menampilkan jenis ikan berdasarkan species
df.groupby(by='Species').describe()
```

```python
df.groupby(by='Species').Weight.describe()
```

```python
label_encoder = LabelEncoder()
df['Label'] = label_encoder.fit_transform(df.Species)
```

```python
df.head()
```

```python
df.Label.value_counts()
```

```python
df.Species.value_counts()
```

```python
X = df.drop(columns=['Weight','Species'])
y = df.Weight

print ("Shape dari X :", X.shape)
print ("Shape dari y :", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=88)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2 = model.score(X_test, y_test)
mse = mean_squared_error(y_pred, y_test)
rmse = np.sqrt(mse)

print("R^2                     : {}".format(r2))
print("Mean Squared Error      : {}".format(mse))
print("Root Mean Squared Error : {}".format(rmse)) #melihat hasil prediksi berat ikan rata - rata
```
